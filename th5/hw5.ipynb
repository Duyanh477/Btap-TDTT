```python
# Nhập số đầu tiên để khởi tạo Max, Min
so = int(input("Nhập số: "))
if so == -1:
    print("Chưa nhập số nào hợp lệ.")
else:
    max_val = min_val = so
    while True:
        so = int(input("Nhập tiếp (nhập -1 để dừng): "))
        if so == -1:
            break
        if so > max_val:
            max_val = so
        if so < min_val:
            min_val = so
    
    print(f"Lớn nhất: {max_val}, Nhỏ nhất: {min_val}")
```

```python
A = int(input("Nhập A: "))
f1, f2 = 1, 1

if A < 1:
    print("Không có số nào.")
elif A == 1:
    print("Số Fibonacci lớn nhất là 1")
else:
    while f1 + f2 <= A:
        f_moi = f1 + f2
        f1 = f2
        f2 = f_moi
    print(f"Số Fibonacci lớn nhất không vượt quá {A} là: {f2}")
```

```python
s = input("Nhập chuỗi: ")
dem_tu = 0
dang_trong_tu = False

for ky_tu in s:
    if ky_tu != ' ':
        if not dang_trong_tu:
            dem_tu += 1
            dang_trong_tu = True
    else:
        dang_trong_tu = False

print(f"Số lượng từ: {dem_tu}")
```

```python
s = input("Nhập chuỗi: ")
chuoi_da_bat_dau = False

for i in range(len(s)):
    if s[i] != ' ':
        chuoi_da_bat_dau = True
        print(s[i], end='') 
    else:
        if chuoi_da_bat_dau:
            break
```

```python
s = input("Nhập chuỗi: ")
ket_qua = ""

for chu in s:
    if chu.isupper():
        ket_qua += chu.lower()
    elif chu.islower():
        ket_qua += chu.upper()
    else:
        ket_qua += chu 

print(f"Chuỗi sau khi đổi: {ket_qua}")
```

```python
s = input("Nhập (vd 3, 12, 15): ")
danh_sach_so = s.split(',')
tong = 0
for x in danh_sach_so:
    tong += int(x.strip()) 

print(f"Tổng: {tong}")
```

```python
s = input("Nhập chuỗi: ")
hoa = thuong = so = 0

for chu in s:
    if chu.isupper():
        hoa += 1
    elif chu.islower():
        thuong += 1
    elif chu.isdigit():
        so += 1

print(f"Chữ hoa: {hoa}, Chữ thường: {thuong}, Chữ số: {so}")
```
