```python
def is_armstrong(x):
    if x < 0: return False
    temp = x
    n = 0
    while temp > 0:
        n += 1
        temp //= 10
        
    temp = x
    tong = 0
    while temp > 0:
        chu_so = temp % 10      
        tong += chu_so ** n     
        temp //= 10             

    return tong == x

x= int(input())
print(is_armstrong(x))
```

```python
def is_perfect_square(x):
    if x < 0: return False
    root = int(x ** 0.5)
    return root * root == x
 
x = int(input())
print(is_perfect_square(x))
```

```python
def doixung(x):
    if x < 0: return False
    temp = x
    so_dao_nguoc = 0
    while temp > 0:
        chu_so_cuoi = temp % 10
        so_dao_nguoc = so_dao_nguoc * 10 + chu_so_cuoi
        temp //= 10
    return x == so_dao_nguoc

x = int(input())
print(doixung(x))
```

```python
def kidieu(x):
    if x <= 0: return False
    
    while x > 9:
        tong_chu_so = 0
        temp = x
        while temp > 0:
            tong_chu_so += temp % 10
            temp //= 10
        x = tong_chu_so 
    return x == 1

x = int(input())
print(kidieu(x))
```
