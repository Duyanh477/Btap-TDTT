```python
def Max_Of_two(a,b) -> int:
    if a > b:
        return a
    elif a < b:
        return b
    if a == b:
        return ("Hòa")

a,b = map(int,input().split())
Max_Of_two(a,b)
```

```python
def nguyen_to(a):
    if a <= 1:
        return"khong co"
    
    for i in range(2,int(a ** 0.5) + 1):
        if a % i == 0:
            return "Khong la nguyen to"
    return "La nguyen to"


a = int(input())
print(nguyen_to(a))
```

```python
def perfect_number(a) -> int:
    tong = 0
    for i in range (1,a):
        if a % i == 0:
            tong += i
    if tong == a:
        return "Hoan hao"
    return"Ko hoan hao"

a = int(input())
print(perfect_number(a))
```

```python
def giaithua(x):
    if x < 0:
        return False
    tich = 1
    for i in range(1,x + 1):
        tich *= i
    return tich

x = int(input())
print(giaithua(x))
```

```python
def compute(a,b,op):
    if op == '+':
        return a + b
    elif op == '-':
        return a - b
    elif op == '*':
        return a * b
    elif op == '/':
        return a / b

a,b = map(float,input().split())
op = input()
print(compute(a,b,op))
```

```python
def tong(x):
    tong = 0
    while x > 0:
        n = x % 10
        tong += n
        x //= 10
    return tong

x = int(input())
print(tong(x))
```
