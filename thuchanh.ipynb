### Problem 4
#### Write a function that returns the count of positive integers k such that 1 <= k < n and k > (n - 1) / 2. (Example: n=9 -> 4).

```python
n = int(input())
def count_k(n):
    count = 0
    ss = (n - 1) / 2
    for k in range(1, n):
        if k > ss:
            count += 1
    return count

print(count_k(n))
```

### Problem 6
#### Write a function that computes the area of a triangle with side lengths a, b, c (use Heron's formula). Raise ValueError if the sides cannot form a triangle.

```python
import math
a = float(input("Nhập số a: "))
b = float(input("Nhập số b: "))
c = float(input("Nhập số c: "))

if (a + b > c) and (a + c > b) and (b + c > a):
    p = (a + b + c) / 2
    s = math.sqrt(p * (p - a) * (p - b) * (p - c))
    print(f"Dien tich tam giac la: {s:.1f}")
else:
    print("ValueError")
```

### Problem 8

#### Preserve case and wrap around within A..Z or a..z.

```python
c = input("Nhập vào một chữ cái: ")

if len(c) == 1 and 'a' <= c <= 'z':
   
        ki_tu_sau = chr(ord(c) + 32)
        print(f"Chữ cái thường liền sau tương ứng: {ki_tu_sau}")
elif len(c) == 1 and 'A' <= c <= 'Z':
        ki_tu_sau = chr(ord(c) + 32)
        print(f"Chữ cái thường liền sau tương ứng: {ki_tu_sau}")
else:
    print("Vui lòng nhập đúng một chữ cái (từ a đến z hoặc A đến Z).")
```

### Problem 9
#### Write a function that converts an uppercase letter to lowercase. If the input is not a single character, raise ValueError.

```python
a = input("Nhap vao mot chu: ")
if len(a) == 1:
    if a.isupper():
        print(f'Day la chu in thuong cua {a.lower()}')
    elif a.islower():
        print(f'Day la chu in thuong {a}')
    else:
        print("Day khong phai la chu cai.")
else:
    print("Vui long nhap vao mot chu cai duy nhat.")
```

### Problem 10
#### Write a function that, given a lowercase letter, returns a string containing that letter repeated 10 times.

```python
char = input()
def repeat_letter_loop(char):
    result = ""
    for i in range(10):
        result += char
    return result
print(repeat_letter_loop(char))
```
