```python
a, b, c = map(float,input().split())

if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        loai = "Tam giác đều"
    elif a == b or b == c or a == c:
        loai = "Tam giác cân"
    elif (a**2 + b**2 == c**2) or (a**2 + c**2 == b**2) or (b**2 + c**2 == a**2):
        loai = "Tam giác vuông"
    else:
        loai = "Tam giác thường"
    
    
    chu_vi = a + b + c
    p = chu_vi / 2
    heron = (p * (p - a) * (p - b) * (p - c))
    dien_tich = heron ** 0.5
    
    print(f"{loai}, Chu vi: {chu_vi}, Diện tích: {dien_tich:.2f}")
else:
    print("Khong la ba canh cua mot tam giac")
```

```python
a, b, c  = map(float,input().split())

if a == 0:
    if b == 0:
        print("Vô số nghiệm" if c == 0 else "Vô nghiệm")
    else:
        nghiem = -c / b
        print(f"x = {nghiem:.2f}")
else:
    delta = b**2 - 4*a*c
    if delta < 0:
        print("Vô nghiệm")
    elif delta == 0:
        nghiem = -b / (2 * a)
        print(f"Nghiệm kép x = {nghiem:.2f}")
    else:
        x1 = (-b + delta ** 0.5) / (2*a)
        x2 = (-b - delta ** 0.5) / (2*a)
        print(f"x1 = {x1:.2f}, x2 = {x2:.2f}")
```
