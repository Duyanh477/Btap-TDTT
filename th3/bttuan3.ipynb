```python
a = int(input("Nhap vao mot so: "))
if a < 100 or a > 999:
    print("Nhap lai di")
else:
    tram = a // 100
    chuc = (a % 100) // 10
    don_vi = a % 10
    dao_nguoc = don_vi * 100 + chuc * 10 + tram
    print(f'số đảo ngược là: {dao_nguoc}')
```

```python
n = float(input())


if n >= 0:
    xuong = int(n)
else:
    if n == int(n): 
        xuong = int(n)
    else:         
        xuong = int(n) - 1


if n == int(n):
    len = int(n)
else:
    if n >= 0:
        len = int(n) + 1
    else:
        len = int(n)


phan_thap_phan = n - xuong
if phan_thap_phan >= 0.5:
    gan_nhat = xuong + 1
else:
    gan_nhat = xuong


print(f"{len} {xuong} {gan_nhat}")
```
