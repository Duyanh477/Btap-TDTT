```python
# Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím 
h = int(input("Nhập độ cao h: "))

for i in range(1, h + 1):
    print(" " * (h - i), end="")
    if i == 1:
        print("*")
    elif i == h:
        print("*" * (2 * i - 1))
    else:
        print("*" + " " * (2 * i - 3) + "*")
```

```python
# Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không. 
# Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28…
a = int(input())
tong = 0
for i in range(1,a):
    if a % i == 0:
        tong += i
if tong == a:
    print(f'{a} is perfect')
else:
    print(f'{a} is imperfect')
```

```python
a = int(input())
b = int(input())
uoc1 = []
uoc2 = []
for i in range(1,a):
    if a % i == 0:
        uoc1.append(i)
for i in range(1,b):
    if b % i == 0:
        uoc2.append(i)

if sum(uoc1) == b and sum(uoc2) == a:
    print(f'{a} and {b} are amicable numbers')
```

```python
n = int(input())
uoc = 2

while n > 1:
    if n % uoc == 0:
        n //= uoc
    else:
        uoc += 1
print(uoc)
```

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

a = int(input())
b = int(input())
tong = 0
for i in range(a, b + 1):
    if is_prime(i):
        tong += i
print(tong)
```

```python
num = int(input("Nhập số nguyên (nhập -1 để dừng): "))

if num == -1:
    print("Bạn chưa nhập số nào hợp lệ.")
else:
    max_val = num
    min_val = num
    
    while True:
        num = int(input("Nhập số tiếp theo: "))
        if num == -1:
            break

        if num > max_val:
            max_val = num    
    
        if num < min_val:
            min_val = num

    print(f"Số lớn nhất là: {max_val}")
    print(f"Số nhỏ nhất là: {min_val}")
```

```python
A = int(input("Nhập A: "))
f1, f2 = 1, 1

while f1 + f2 <= A:
    # f3 = f1 + f2
    # f1 = f2
    # f2 = f3
    f1, f2 = f2, f1 + f2

print(f"Số Fibonacci lớn nhất <= A là: {f2}")
```

```python
n = int(input())
nt = [2, 3, 5, 7]
dem = 0
while n > 0:
    so = n % 10
    if so in nt:
        dem += 1
    n //= 10 
print(dem)
```

```python
n = int(input("Nhập n: "))
is_power_of_3 = True

if n <= 0:
    is_power_of_3 = False
else:
    while n > 1:
        if n % 3 != 0:
            is_power_of_3 = False
            break
        n //= 3

if is_power_of_3:
    print("Dạng 3^k")
else:
    print("Không phải dạng 3^k")
```
