```python
a, b = map(int,input().split())
if a > b or a <= 0 or b <= 0:
    print(f'Nhap lai')
else:
    a = max(a, 2)
    tong = 0
    for i in range(a,b +1):
        for j in range(2,int(i**0.5) + 1):
            if i % j == 0:
                break
        else:
            tong = tong + i
    print(tong)
```

```python
n = int(input())
if n < 2:
    print(f'Nhap lai')
else:
    nguyenlon = 0
    for i in range(2, int(n**0.5)+1):
        while n % i == 0:
            nguyenlon = i
            n = n // i

    if n > 1:
        nguyenlon = n
    print(nguyenlon)
```
