```python
# Hãy tính phần
# diện tích còn để trồng cây
d = int(input("Chieu dai: "))
r = int(input("Chieu rong: "))
pi = 3.14
tron = pi * (r / 2) ** 2
hcn = d * r
con_lai = hcn - tron
print(f'Dien tich con lai la: {con_lai:.2f}')
```

```python
a = input("Nhap vao mot chu: ")
if len(a) == 1:
    if a.isupper():
        print(f'Day la chu in thuong {a.lower()}')
    elif a.islower():
        print(f'Day la chu in hoa {a.upper()}')
    else:
        print("Day khong phai la chu cai.")
else:
    print("Vui long nhap vao mot chu cai duy nhat.")
```

```python
c = input("Nhap vao mot ki tu: ")
if len(c) == 1:
    if c.isalpha():
       print(f'{c} la chu cai')
    else:
        print(f'{c} khong phai la chu cai')
elif len(c) != 1:
    print("Vui long nhap vao mot ki tu duy nhat.")
```

```python
c = input("Nhập vào một chữ cái hoa: ")

if len(c) == 1 and 'A' <= c <= 'Z':
    if c == 'A':
        print("Trường hợp đặc biệt: Chữ cái thường tương ứng là 'a'.")
        print("Kí tự liền trước 'a': z")   
    else:
        ki_tu_truoc = chr(ord(c) + 31)
        print(f"Chữ cái thường liền trước tương ứng: {ki_tu_truoc}")
else:
    print("Vui lòng nhập đúng một chữ cái hoa (từ A đến Z).")
```

```python
import math
a = float(input("Nhập số a: "))
b = float(input("Nhập số b: "))
c = float(input("Nhập số c: "))

if (a + b > c) and (a + c > b) and (b + c > a):
    p = (a + b + c) / 2
    s = math.sqrt(p * (p - a) * (p - b) * (p - c))
    print(f"Dien tich tam giac la: {s:.1f}")
else:
    print("Khong phai 3 canh cua tam giac")
```

```python
ten_chu_ho = input("Ten chu ho: ")
chi_so_truoc = int(input("Chi so thang truoc: "))
chi_so_nay = int(input("Chi so thang nay: "))

kwh = chi_so_nay - chi_so_truoc
tien_dien = 0

if kwh <= 50:
    tien_dien = kwh * 1984
elif kwh <= 100:
    tien_dien = 50 * 1984 + (kwh - 50) * 2050
elif kwh <= 200:
    tien_dien = 50 * 1984 + 50 * 2050 + (kwh - 100) * 2380
elif kwh <= 300:
    tien_dien = 50 * 1984 + 50 * 2050 + 100 * 2380 + (kwh - 200) * 2998
elif kwh <= 400:
    tien_dien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + (kwh - 300) * 3350
else:
    tien_dien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + 100 * 3350 + (kwh - 400) * 3460

tong_tien_co_vat = tien_dien * 1.08

tien_phai_tra = round(tong_tien_co_vat)

print(f"Ho va ten: {ten_chu_ho}")
print(f"Tien phai tra la: {tien_phai_tra}")
```
